```python
import numpy as np
import pandas as pd
import matplotlib as mlt
```

```python
data = pd.read_csv('mushrooms.csv');
data = data.sample(frac=1)
data.head()
```

```python
splitrow = round(0.75 * data.shape[0])
datalatih = data.iloc[:splitrow,:]
datauji = data.iloc[splitrow:,:]
```

```python
kelas = pd.crosstab(index = datalatih['class'],columns="count")
probabilityKelas = kelas/kelas.sum()
probabilityKelas
```

```python
a = pd.crosstab(datalatih["cap-shape"],datalatih["class"]) #lihat bentul data 
a=a+1#laplace correction
a
```

```python
matrixproball = {}
for i in range(1,len(kolom)):
    crosstab = pd.crosstab(datalatih[kolom[i]],datalatih['class'])
    crosstab = crosstab+1;
    for a in range(crosstab.shape[1]):
        crosstab.iloc[:,a] = crosstab.iloc[:,a]/crosstab.iloc[:,a].sum()
    matrixproball[kolom[i]]=crosstab
```

```python
classe = probabilityKelas['count']['e'];
classp = probabilityKelas['count']['p'];
matrixproball[datauji.columns[1]]['e'][datauji[datauji.columns[1]].iloc[0]]
```

```python
y_predict = [];
for i in range(datauji.shape[0]):
    classe = probabilityKelas['count']['e'];
    classp = probabilityKelas['count']['p'];
    for j in range(1,len(datauji.columns)):
        classe = classe*matrixproball[datauji.columns[j]]['e'][datauji[datauji.columns[j]].iloc[i]]
        classp = classp*matrixproball[datauji.columns[j]]['p'][datauji[datauji.columns[j]].iloc[i]]
        #print(datatest.columns[j])
        #print(datatest[datatest.columns[j]].iloc[i])
        #print(matrixproball[datatest.columns[j]]['e'][datatest[datatest.columns[j]].iloc[0]])
        #print(matrixproball[datatest.columns[j]]['p'][datatest[datatest.columns[j]].iloc[0]])
    if(classe>classp):
        y_predict.append('e')
    else:
        y_predict.append('p')
```

```python
y_actual = datauji['class'].tolist()
```

```python
hasil = {'y_predict':y_predict,
         'y_actual' :y_actual}
```

```python
hsl = pd.DataFrame(hasil,columns=['y_actual','y_predict'])
```

```python
confmat = pd.crosstab(hsl['y_actual'],hsl['y_predict'],rownames=['actual'],colnames=['predict'])
confmat
```

```python
datauji[datauji['class']=='e']['class'].count()
```

```python
recalle = (confmat['e']['e']/confmat.iloc[0].sum())*100
recallp = confmat['p']['p']/confmat.iloc[1].sum()*100
prece = confmat['e']['e']/confmat.iloc[:,0].sum()*100
precp = confmat['p']['p']/confmat.iloc[:,1].sum()*100
```
